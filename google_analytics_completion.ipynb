```python
import numpy as np
import pandas as pd
import csv
import datetime
import matplotlib.pyplot as plt
```

```python
def readGoogleAnalytics(filename):
    header = ['category','date','OS','browser','version','type','expression','localization','local time','remote time','total time','status']
    df = pd.DataFrame(columns=header)
    i = 0
    with open(filename, "r") as csvfile:
        reader = csv.DictReader(filter(lambda row: row[0] != '#', csvfile))
        for rows in reader:
            row = list(rows.values())[0]
            words = row[0].split(sep=',')
            if (len(header) == len((words))):
                df.loc[i] = words
                i = i+1
    return df
```

```python
def readGoogleAnalytics(file_name):
    d = open(file_name)
    df = pd.DataFrame(columns=('expr','kind','local','combined','local_time','remote_time','total_time','status'))
    i = 0
    for line in d:
        write_line = False
        words = line.split(sep=',')
        expr = ''
        subs = "expr="
        if any(subs in s for s in words):
            expr = [s for s in words if subs in s]
            expr = ''.join(expr)
            expr = expr[5:]
            write_line = True
        kind =''
        subs = "kind="
        if any(subs in s for s in words):
            kind = [s for s in words if subs in s]
            kind = ''.join(kind)
            kind = kind[6:]
        subs = "local"
        if (any(subs in s for s in words)) and (not(any("time" in s for s in words))):
            local = True
        else:
            local = False
        subs = "combined"
        if any(subs in s for s in words):
            combined = True
        else:
            combined = False
        subs = "local time="
        if any(subs in s for s in words):
            local_time = [s for s in words if subs in s]
            local_time = ''.join(local_time)
            local_time = local_time[12:]
            local_time = local_time[:-2]
        else:
            local_time = 0
        subs = "remote time="
        if any(subs in s for s in words):
            remote_time = [s for s in words if subs in s]
            remote_time = ''.join(remote_time)
            remote_time = remote_time[13:]
            remote_time = remote_time[:-2]
        else:
            remote_time = 0
        subs = "total="
        if any(subs in s for s in words):
            total_time = [s for s in words if subs in s]
            total_time = ''.join(total_time)
            total_time = total_time[7:]
            total_time = total_time[:-2]
        else:
            total_time = 0
        status =''
        subs = "status="
        if any(subs in s for s in words):
            status = [s for s in words if subs in s]
            status = ''.join(status)
            status = status[8:]
        if write_line:
            local_time = int(total_time) - int(remote_time)
            df.loc[i] = [expr,kind,local,combined,local_time,remote_time,total_time,status]
            i = i+1
    return df
```
